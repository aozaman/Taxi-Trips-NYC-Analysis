```python
#Data (install packages)
import pandas as pd
import numpy as np
```

```python
#Read the data
data = pd.read_csv('yellow_tripdata.csv')
```

```python
#Getting to know the data
data.head()
```

```python
data.info()
```

```python
#Are there any missing values?
data.isnull().sum()
```

```python
#Cleaning the dataset
data.dropna(inplace = True)
```

```python
#Checking to see if the null values have been removed
data.isnull().sum()
```

```python
#Extracting two columns
location=data[['PULocationID','DOLocationID']]
print(location)
```

```python
#Using kMeans clustering to create 5 boroughs/clusters for NYC
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters = 5, init = 'k-means++', random_state = 42)
X=location
y_kmeans = kmeans.fit_predict(X)
```

```python
y_kmeans
```

```python
data['Clusters']=y_kmeans
```

```python
data.head(20)
```

```python
#Converting from string to date
from datetime import datetime
data['tpep_pickup_datetime']= pd.to_datetime(data['tpep_pickup_datetime'])
data['tpep_dropoff_datetime']= pd.to_datetime(data['tpep_dropoff_datetime'])
data.info()
```

```python
#Calculating trip duration
data['TripDuration']= (data['tpep_dropoff_datetime'] - data['tpep_pickup_datetime'])/np.timedelta64(1,'s') 
data.head()
```

```python
X = data[['trip_distance','Clusters','TripDuration']]
y = data.total_amount
X.shape
```

```python
import sklearn.linear_model
import matplotlib.pyplot as plt
#Calling the model
model = sklearn.linear_model.LinearRegression()

#Fit the model on the data
model.fit(X,y)

#Predicting the ys
y_pred = model.predict(X) #Use X to predict y
print(y_pred)
```

```python
#Accuracy
r_sq = model.score(X,y)
print(r_sq)
```
